# Salary Dataset Exploration
This notebook explores the relationship between years of experience and salary, performs basic cleaning, and visualizes key patterns.

```python
# Load libraries and dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('data/raw/Salary_Data.csv')
df.head()
```

```python
# Basic cleaning
df = df.dropna()
df.describe()
```

```python
# Scatter plot: YearsExperience vs Salary
sns.set(style='whitegrid')
ax = sns.scatterplot(data=df, x='YearsExperience', y='Salary')
ax.set_title('Years of Experience vs Salary')
plt.show()
```
